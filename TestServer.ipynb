```python
import requests
import json
import pandas as pd
from io import StringIO

URL_ITEM = 'http://127.0.0.1:8000/predict_item'
URL_ITEMS = 'http://127.0.0.1:8000/predict_items'
URL_CSV = 'http://127.0.0.1:8000/csv_content'
```

```python
!start uvicorn server:app --reload
```

```python
def get_response(url, data):
    return json.loads(requests.post(url, json=data).text)
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/Murcha1990/MLDS_ML_2022/main/Hometasks/HT1/cars_test.csv').dropna()
objects = [{key: value for key, value in zip(df.columns.tolist(), row)} for row in df.values.tolist()]
df
```

```python
one_object = objects[0]
one_object
```

```python
predict_one = get_response(URL_ITEM, one_object)
predict_one
```

```python
predict_all = get_response(URL_ITEMS, objects)
predict_all
```

```python
len(predict_all)
```

```python
csv_buffer = StringIO()
df.to_csv(csv_buffer, index=False)
csv_str = get_response(URL_CSV, {'content': csv_buffer.getvalue()})
df_result = pd.read_csv(StringIO(csv_str['content']))
df_result
```
